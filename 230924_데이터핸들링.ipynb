```python
# 8. 데이터 프레임 컬럼확인, 컬럼명변경, 컬럼타입변경
df = pd.read_csv('data/iris.csv')
print(type(df.columns))
print(df.columns)
df.columns = ['a_a', 'b_b', 'c_c','d_d','e_e']
print(df.columns)
df.columns = df.columns.str.replace('_','')
print(df.columns)
df.rename(columns={'bb':'b_b'}, inplace = True)
print(df.columns)
print(df.info())
df['aa'] = df['aa'].astype('int') # int, str, float
print(df.info())
df = df.convert_dtypes()
df.info()
```

```python
# 9. 데이터 프레임 행/열 선택
# loc : 레이블 기반 인덱서, iloc : 위치 기반 인덱서
df = pd.read_csv('data/iris.csv')
print(df)
loc = df.loc[1:4, ['sepal length', 'sepal width']] # 행(닫힌구간) / 열(컬럼명지정)
print(loc)
iloc = df.iloc[1:4, 1:4] # 행(반열린구간) / 열(반열린구간)
print(iloc)
iloc = df.iloc[[1,4], [1,4]] #행(지정) / 열(지정)
print(iloc)
```

```python
# 10. 데이터 프레임 행 추가/삭제
# 행 추가
import pandas as pd
df = pd.read_csv('data/iris.csv')
list = [[99.0,98.0,97.0,96.0, 'sample']]
sample = pd.DataFrame(list, columns = df.columns)
print(sample)
#df = df.append(sample) 
df = pd.concat([df,sample], axis = 0) # pandas에서 append 미추천?
df.reset_index(drop = True, inplace = True) # drop = True : 기존 index 삭제 (default = False)
print(df)
# 행 삭제
df.drop(150,inplace = True) # 인덱스 150번째 행 삭제
print(df)
df.drop([0,3], inplace = True) # 인덱스 0, 3번째 행 삭제
print(df)
df.drop(df.index[1:3], inplace = True) # 1,2번째 행 삭제(반열림구간) > 인덱스가 아니라 실제 행 기준으로 삭제
df
# 조건에 따른 행 삭제?

# 결측치가 있는 행 삭제
import numpy as np
df = pd.read_csv('data/iris.csv')
list = [[99.0, np.nan,1,1,'nan_sample']]
sample = pd.DataFrame(list, columns = df.columns)
df = pd.concat([df, sample], axis = 0)
print(df)
df.dropna(inplace = True)
print(df)

# 중복행 추출 및 제거
df_temp = df[df.duplicated() == True]
print(df_temp)
df_temp = df_temp.drop_duplicates(keep = 'last')
print(df_temp)
df.drop(df[df.duplicated()].index, axis = 0, inplace = True)
df
```
