```python
# 10. 데이터 프레임 행 추가/삭제
# 행 추가
import pandas as pd
df = pd.read_csv('data/iris.csv')
list = [[99.0,98.0,97.0,96.0, 'sample']]
sample = pd.DataFrame(list, columns = df.columns)
print(sample)
#df = df.append(sample) 
df = pd.concat([df,sample], axis = 0) # pandas에서 append 미추천?
df.reset_index(drop = True, inplace = True) # drop = True : 기존 index 삭제 (default = False)
print(df)
# 행 삭제
df.drop(150,inplace = True) # 인덱스 150번째 행 삭제
print(df)
df.drop([0,3], inplace = True) # 인덱스 0, 3번째 행 삭제
print(df)
df.drop(df.index[1:3], inplace = True) # 1,2번째 행 삭제(반열림구간) > 인덱스가 아니라 실제 행 기준으로 삭제
df
# 조건에 따른 행 삭제?

# 결측치가 있는 행 삭제
import numpy as np
df = pd.read_csv('data/iris.csv')
list = [[99.0, np.nan,1,1,'nan_sample']]
sample = pd.DataFrame(list, columns = df.columns)
df = pd.concat([df, sample], axis = 0)
print(df)
df.dropna(inplace = True)
print(df)

# 중복행 추출 및 제거
df_temp = df[df.duplicated() == True]
print(df_temp)
df_temp = df_temp.drop_duplicates(keep = 'last')
print(df_temp)
df.drop(df[df.duplicated()].index, axis = 0, inplace = True)
df
```

```python
# 12. 조건에 따른 행/열 선택
import pandas as pd
df = pd.read_csv('data/titanic.csv')
print(df)
# 생존자 추출
surv = df['survived'] == 1
df[surv]

# 생존자 and 요금 30달러 이상 and First class 에서 fare만 추출
surv  = df['survived'] == 1
fare  = df['fare'] >= 30
first = df['class'] == 'First'
df[surv & fare & first][['fare']]

# 조건에 맞는 행 삭제
df.drop(df[surv & fare & first].index , axis = 0, inplace = True)
df
```

```python
# 14. np.where(), np.select() : 필드에 접근하고 수정
import pandas as pd
import numpy as np
df = pd.read_csv('data/titanic.csv')
# np.where()
df['temp_class_3'] = 'n'
df.loc[df['class'] == 'First', 'temp_class_2'] = 'y'
df['temp_class_2'] = 'n'
df.iloc[df[df['class'] == 'First'].index, len(df.columns)-1] = 'y'
df['temp_class_1'] = np.where(df['class'] == 'First', 'y', 'n')

# np.select() : 조건 목록과 선택목록의 길이는 동일

con_list = [df['age'] >= 19,
            (df['age'] >=8) & (df['age'] <19),
            df['age'] <8
           ]
sel_list = ['성인', '학생', '유아']

df['temp_age'] = np.select(con_list, sel_list, default = 'not_specified')
df[df['temp_age'] == '성인']
df[df['temp_age'] == '학생']
df[df['temp_age'] == '유아']
```

```python
# 19. merge : 두 데이터 프레임에 존재하는 고유값(key)를 기준으로 병합
import pandas as pd
import numpy as np

id_name = pd.DataFrame(
                       {'id' : ['c01', 'c02', 'c03', 'c04', 'c05', 'c06', 'c07'],
                        'last_name' : ['Lee', 'Kim', 'Choi', 'Park','Lim','Bae','Kim']   
                       }
                      )
id_number = pd.DataFrame(
                         { 'id' : ['c03', 'c04', 'c05', 'c06', 'c07', 'c08', 'c09'],
                           'last_name' : ['Choi', 'Park','Lim','Bae','Kim',np.nan,np.nan],      
                           'number' : [3, 1, 0, 7, 3, 4, 1]  
                         }
                        )

#print(id_name)
#print(id_number)

df_inner = pd.merge(id_name, id_number, how = 'inner', on = 'id')
print(df_inner)
df_inner_two = pd.merge(id_name, id_number, how = 'inner', on = ['id', 'last_name'])
print(df_inner_two)
df_left  = pd.merge(id_name, id_number, how = 'left', on = 'id')
df_right = pd.merge(id_name, id_number, how = 'right', on = 'id')
df_outer = pd.merge(id_name, id_number, how = 'outer', on = 'id')
```

```python
# 20. 특정 컬럼을 기준으로 데이터를 그룹지어 집계함수 적용
# sum, count, mean, median, var, std, min, max, unique, nunique, prod, first, last

import pandas as pd
df = pd.read_csv('data/titanic.csv')

# 'sex','class' 을 기준으로 데이터를 그룹지어 'age' 와 'fare'의 평균함수 추출
df.groupby(['sex', 'class'])[['fare','age']].mean().reset_index()

# 다중 통계량
df_tmp = df.groupby(['sex', 'class'])[['fare','age']].agg(['mean', 'median']).reset_index()
df_tmp

# 컬럼별 다중 통계량
df_tmp_2 = df.groupby(['sex','class'])[['fare','age']].agg({'fare' : 'mean', 'age' : ['mean','median']}).reset_index()
df_tmp_2
print(df_tmp_2)
print(df_tmp_2.columns.to_flat_index())

# 컬럼 평탄화 작업
def flat_cols(df):
    df.columns = [' / '.join(x) for x in df.columns.to_flat_index()]
    return df
df_tmp_2 = flat_cols(df_tmp_2)
df_tmp_2

```
