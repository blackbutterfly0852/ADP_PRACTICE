```python
# 20. 특정 컬럼을 기준으로 데이터를 그룹지어 집계함수 적용
# sum, count, mean, median, var, std, min, max, unique, nunique, prod, first, last

import pandas as pd
df = pd.read_csv('data/titanic.csv')

# 'sex','class' 을 기준으로 데이터를 그룹지어 'age' 와 'fare'의 평균함수 추출
df.groupby(['sex', 'class'])[['fare','age']].mean().reset_index()

# 다중 통계량
df_tmp = df.groupby(['sex', 'class'])[['fare','age']].agg(['mean', 'median']).reset_index()
df_tmp

# 컬럼별 다중 통계량
df_tmp_2 = df.groupby(['sex','class'])[['fare','age']].agg({'fare' : 'mean', 'age' : ['mean','median']}).reset_index()
df_tmp_2
print(df_tmp_2)
print(df_tmp_2.columns.to_flat_index())

# 컬럼 평탄화 작업
def flat_cols(df):
    df.columns = [' / '.join(x) for x in df.columns.to_flat_index()]
    return df
df_tmp_2 = flat_cols(df_tmp_2)
df_tmp_2
```

```python
# 21. 범주형 변수의 도수분포표, 이원분할표
import pandas as pd
df = pd.read_csv('data/titanic.csv')
se_survived = df['survived'].value_counts()
df_survived = se_survived.to_frame()
df_survived

# 기준이 두 개일 경우
pd.crosstab(df['class'], df['survived'])

# 범주형 변수에 대한 전체 상대도수 비율
pd.crosstab(df['class'], df['survived'], normalize=True)

# 범주형 변수에 대한 행 상대도수 비율
pd.crosstab(df['class'], df['survived'], normalize='index')

# 범주형 변수에 대한 열 상대도수 비율
pd.crosstab(df['class'], df['survived'], normalize ='columns')
```

```python
# 22. 데이터 재구조화 (melt, pivot_table)
import pandas as pd
import numpy as np
df = pd.read_csv('data/Fruits.csv')
# melt
df_melt = pd.melt(df, id_vars = ['Fruit'], value_vars =['Sales', 'Expenses', 'Profit'], var_name = 'subjects', value_name = 'value', ignore_index = True)
df_melt

# pivot_table : 자동 Group by + mean (default), margins = True (전체평균출력)
df_pivot = pd.pivot_table(df_melt, index = 'Fruit', columns = 'subjects', values = 'value', margins = True)
df_pivot = pd.pivot_table(df_melt, index = ['Fruit'], columns = ['subjects'], values = ['value'], aggfunc = ['mean', 'median'], margins = [True])
df_pivot

# 컬럼 평탄화 작업
def flat_cols(df):
    df.columns = [' / '.join(x) for x in df.columns.to_flat_index()]
    return df
df_pivot = flat_cols(df_pivot).reset_index()
df_pivot
```

```python
# 27. lambda
# lambda 간단한 함수 개념으로, map(lambda)를 사용할 때는 Series(원소) 형식 필요
# 대체원소 -> 원소.func()
import pandas as pd
df = pd.read_csv('data/6.세종시_오피스텔(매매)_실거래가.csv')
df_split = df['시군구'].str.split(' ', expand = True)
df_split

# lambda로 split
se_split_lambda = df['시군구'].map(lambda x : x.split(' '))
df_from_se_split_lambda = pd.DataFrame(columns = ['시','군','구'])
df_from_se_split_lambda

temp_df = pd.DataFrame([se_split_lambda[0]], columns = ['시','군'])
pd.concat([df_from_se_split_lambda, temp_df], axis = 0)

# 리스트를 [] 감싸서 데이터프레임으로 생성 후 pd.concat() 실행
for i in se_split_lambda:
    if len(i) == 2:
        temp_df = pd.DataFrame([i], columns = ['시','군'])
        df_from_se_split_lambda = pd.concat([df_from_se_split_lambda, temp_df], axis = 0)
    else:
        temp_df = pd.DataFrame([i], columns = ['시','군','구'])
        df_from_se_split_lambda = pd.concat([df_from_se_split_lambda, temp_df], axis = 0)

df_from_se_split_lambda.reset_index(inplace = True)
df_from_se_split_lambda[df_from_se_split_lambda['구'].notnull()]


# 타입변경 후 문자열 자르기
df['계약년월'] = df['계약년월'].astype(str)[0:4]
df.info()
df['계약년월'].map(lambda x : str(x)[0:4])
```

```python
# 28. 날짜/시간 다루기
from datetime import datetime
# 기본
print(datetime.today())
print(datetime.today().year)
print(datetime.today().month)
print(datetime.today().day)
print(datetime.today().hour)
print(datetime.today().minute)
print(datetime.today().second)
print(datetime.today().weekday()) # 0 : 월요일 ~ 6 : 일요일
print('----------')

# strftime : datetime > str
today = datetime.today().strftime('%Y %y 년 %m 월 %d 일 %H %I 시 %M 분 %S 초 %f 마이크로초 %A %a 요일')
print(today)
print(type(today))
today_split = today.split(' ')
print(len(today_split))
print(today_split[0])
print('----------')

# strptime : str > datetime 
today = datetime.today().strftime('%Y %m %d %H %M %S %f %A %a') # 월은 소문자 m, 분은 대문자 M > 월, 일, 마이크로초만 소문자
today_datetime = datetime.strptime(today, '%Y %m %d %H %M %S %f %A %a') # 요일 생략
print(today)
print(type(today))
print(today_datetime)
print(type(today_datetime))
print('----------')

# 현재 날짜로부터 1일 후의 날짜를 구해라
from datetime import timedelta
today = datetime.today()
print(today)
print(today + timedelta(days = 1, weeks = 1, hours = 1, minutes = 1, seconds = 1)) # days, weeks 동시 적용
print('----------')

# 실습
import pandas as pd
from datetime import datetime
df = pd.read_csv('data/구매내역.csv')
# int > str
df['datetime'] = df['datetime'].astype(str)

# pd.to_datetime : DataFrame 내 str > datetime
df['str_to_datetime'] = pd.to_datetime(df['datetime'], format = '%Y %m %d')
print(df)
print('----------')

# DataFrame 내 datetime 형식의 컬럼 내 특정 년/월/일이 존재?
print(df[df['str_to_datetime'].dt.year == 2014]) #'2014'는 문자열이기 때문에 조회 X
print('----------')
print(df[df['str_to_datetime'].dt.month == 7])
print('----------')
print(df[df['str_to_datetime'].dt.day == 7])
print('----------')
print(df[df['str_to_datetime'].dt.month.isin([5,6])])
print('----------')

# dt.strftime : DataFrame 내 datetime > str
df['datetime_to_str'] = df['str_to_datetime'].dt.strftime('%Y-%m-%d')
print(df)
print('----------')

# lambda
# DataFrame 내 str > datetime
df['lambda_str_to_datetime'] = df['datetime_to_str'].map(lambda x : datetime.strptime(x, '%Y-%m-%d'))

# DataFrame 내 datetime > str
df['lambda_datetime_to_str'] = df['str_to_datetime'].map(lambda x : datetime.strftime(x, '%Y-%m-%d'))
df.info()
```
