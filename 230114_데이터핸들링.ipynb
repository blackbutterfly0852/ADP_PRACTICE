```python
# 6. 데이터 프레임 인덱스 확인/추가/리셋
df = pd.read_csv('data/iris.csv')
# sepal : 꽃받침, petal : 꽃잎
# drop = True가 기본 > 인덱스로 설정시 컬럼에 남아있지 않냐?(drop = True), 컬럼에 남아있냐(drop = False)
# sit, rif로 기본
df.set_index('target', inplace = True, drop = False) # 특정 컬럼을 인덱스로 사용, inplace = True : 별도의 대입없이 바로 사용
print(df)
print('----------')
df.set_index('petal length', inplace = True) # petal length로 인덱스를 재지정시, 위의 petal length 인덱스는 자동 삭제
print(df)
print('----------')
df.set_index('petal width', inplace = True) # petal width로 인덱스를 재지정시, 위의 petal length 인덱스는 자동 삭제
print(df)
print('----------')
# drop = False가 기본, 직전 인덱스가 컬럼에 남아있냐(drop = False), 컬럼에 남아있지 않냐?(drop = True)
df.reset_index(inplace = True) # 기본 인덱스로 설정, petal width 인덱스는 컬럼으로 이동(drop = False)
print(df.index)
df
```

```python
# 9. 데이터 프레임 행/열 선택
# loc : 레이블 기반 인덱서, iloc : 위치 기반 인덱서
df = pd.read_csv('data/iris.csv')
print(df)
df_loc = df.loc[1:4,['sepal length' ,'petal width']] # 4행(닫힌구간), 2열
print(df_loc)
# 참고
# 1 < x < 4, (1,4) : 열린구간
# 1 <= x <= 4, [1,4] : 닫힌구간
# 1 <= x < 4 or 1 < x <= 4, [1,4) or (1:4] : 반열린구간(반닫힌구간) > 보통 1 <= x < 4
df_iloc = df.iloc[1:4, [0,3]] # 3행(반열린구간), 2열
print(df_iloc)
```

```python
# 10. 데이터 프레임 행 추가/삭제
# 10-1) 행 삭제
import pandas as pd
df = pd.read_csv('data/iris.csv')
list = [[99.0, 98.0, 97.0, 96.0,'sample']] # 행으로 만들기 위해 대괄호 2개
sample = pd.DataFrame(list, columns = df.columns)
print(sample)
print('----------')
df = df.append(sample)
print(df) # append 직후 index 재설정 필요
print('----------')
df.reset_index(drop = True, inplace = True)
print(df) # append 직후 index 재설정 필요
print('----------')
df = pd.concat([df,sample], axis = 0) # pandas에서 append 미추천
df.reset_index(drop = True, inplace = True)
df
```

```python
# 12. 조건에 따른 행/열 선택
import pandas as pd
df = pd.read_csv('data/titanic.csv')
# 12-1) 생존자 추출
df[df['survived'] == 1]
```

```python
# 12-2) 생존자 and 요금 30달러 이상 and First class
df[(df['survived'] == 1) & (df['fare'] >= 30) & (df['class'] == 'First')]
```

```python
# 13. 조건에 따른 열 추가/변경
# 13-1) 생존자 and 요금 30달러이상 and First class > temp 열 추가 후 '생존'
df.loc[(df['survived'] == 1) & (df['fare'] >= 30) & (df['class'] == 'First'), # 행
        'temp' # 새로운 컬럼 추가
      ] = '생존, 30달러이상, First_class'
df
```

```python
# 13-2) temp2 생성 후 기본값 True, 생존자 and 요금 30달러이상 and First class > False
# 데이터 프레임에 직접 접근할 때에는 loc 사용 필요
df['temp2'] = True
df.loc[(df['survived'] == 1) & (df['fare'] >= 30) & (df['class'] == 'First'), # 행
        'temp2' # 열
      ] = False
df
```

```python
# 13-2) temp NaN or temp2 True 행 삭제
df.drop(df[(df['temp'].isnull() == True) & (df['temp2'] == True)].index, axis = 0, inplace = True)
df
```

```python
# 13-3) temp, temp2 열 삭제
df.drop(['temp', 'temp2'], axis = 1, inplace = True)
df.reset_index(drop = True, inplace = True)
df
```

```python
# 14. np.where(), np.select()
import numpy as np
import pandas as pd
df = pd.read_csv('data/titanic.csv')
df
# 14-1) np.where(condition, True, False) : 데이터 프레임 혹은 리스트 내 필드에 접근하고 수정( == loc | iloc * 2)
# First class이면, temp컬럼에 'Yes' 반환 아니면 'No' 반환
df['temp_where'] = np.where(df['class'] == 'First', 'Yes', 'No')

df['temp_loc'] = 'No'
df.loc[df['class'] == 'First', 'temp_loc'] = 'Yes'

df['temp_iloc'] = 'No'
df.iloc[ [df[df['class'] == 'First'].index] , 13] = 'Yes'

df
```

```python
# 14-2) np.select() : 데이터 프레임 혹은 리스트 내 "필드(열)"에 접근하고 수정
# "조건 목록"과 "선택 목록"을 입력으로 받아서 조건에 따라 선택 목록의 요소로 구성된 배열 반환 > 조건 목록과 선택목록의 길이는 동일해야 한다.
# 두 개이상의 조건이었을 때 Pandas의 주어진 조건을 기반으로 DataFrame의 열을 생성 가능.
condition_list = [
    (df['age'] >= 19) ,
    (df['age'] >= 8) & (df['age'] <19),
    (df['age'] < 8)
]

choice_list = ['성인', '초중고등학생', '미취학아동']

df['temp_select'] = np.select(condition_list, choice_list, default = 'Not_spcified')
df[df['temp_select'] == '초중고등학생']
```
