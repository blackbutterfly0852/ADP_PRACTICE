```python
# 14. np.where(), np.select()
import numpy as np
import pandas as pd
df = pd.read_csv('data/titanic.csv')
df
# 14-1) np.where(condition, True, False) : 데이터 프레임 혹은 리스트 내 필드에 접근하고 수정( == loc | iloc * 2)
# First class이면, temp컬럼에 'Yes' 반환 아니면 'No' 반환
df['temp_where'] = np.where(df['class'] == 'First', 'Yes', 'No')

df['temp_loc'] = 'No'
df.loc[df['class'] == 'First', 'temp_loc'] = 'Yes'

df['temp_iloc'] = 'No'
df.iloc[ [df[df['class'] == 'First'].index] , 13] = 'Yes'

df
```

```python
# 14-2) np.select() : 데이터 프레임 혹은 리스트 내 "필드(열)"에 접근하고 수정
# "조건 목록"과 "선택 목록"을 입력으로 받아서 조건에 따라 선택 목록의 요소로 구성된 배열 반환 > 조건 목록과 선택목록의 길이는 동일해야 한다.
# 두 개이상의 조건이었을 때 Pandas의 주어진 조건을 기반으로 DataFrame의 열을 생성 가능.
condition_list = [
    (df['age'] >= 19) ,
    (df['age'] >= 8) & (df['age'] <19),
    (df['age'] < 8)
]

choice_list = ['성인', '초중고등학생', '미취학아동']

df['temp_select'] = np.select(condition_list, choice_list, default = 'Not_spcified')
df[df['temp_select'] == '초중고등학생']
```

```python
# 19. merge : 두 데이터 프레임에 존재하는 고유값(key)를 기준으로 병합
# 고유값(key)의 명칭은 동일해야 됨.
# pd.merge(df_left, df_right, how = 'inner', 'left', 'right', 'outer', on = None) 
import pandas as pd
import numpy as np
id_name = pd.DataFrame(
                       {'id' : ['c01', 'c02', 'c03', 'c04', 'c05', 'c06', 'c07'],
                        'last_name' : ['Lee', 'Kim', 'Choi', 'Park','Lim','Bae','Kim']   
                       }
                      )
id_number = pd.DataFrame(
                         { 'id' : ['c03', 'c04', 'c05', 'c06', 'c07', 'c08', 'c09'],
                           'last_name' : ['Choi', 'Park','Lim','Bae','Kim',np.nan,np.nan],      
                           'number' : [3, 1, 0, 7, 3, 4, 1]  
                         }
                        )

inner = pd.merge(id_name, id_number, how = 'inner', on = 'id')
print(inner)
print('----------')
left = pd.merge(id_name, id_number, how = 'left', on = 'id')
print(left)
print('----------')
right = pd.merge(id_name, id_number, how = 'right', on = 'id')
print(right)
print('----------')
outer = pd.merge(id_name, id_number, how = 'outer', on = 'id')
print(outer)
print('----------')
inner_on_two = pd.merge(id_name, id_number, how = 'inner', on = ['id', 'last_name']) # 두 개 컬럼 조인
print(inner_on_two)
```

```python
# 20. 특정 컬럼을 기준으로 데이터를 그룹지어 집계함수 적용
import pandas as pd
df = pd.read_csv('data/titanic.csv')

# 20-1) 'sex','class' 을 기준으로 데이터를 그룹지어 평균함수 적용 > 숫자형 컬럼 제외 추출
groupby_mean = df.groupby(['sex','class']).mean().reset_index() # reset_index(drop = True) 시 int index 변경 후 'sex', 'class' 컬럼 삭제
groupby_mean

# 20-2) 'sex','class' 을 기준으로 데이터를 그룹지어 평균함수 적용 후 'age' 와 'fare' 추출
groupby_mean_column = df.groupby(['sex','class']).mean()[['age','fare']] # 두 개 이상 컬럼 선택시 대괄호 2개 주의
groupby_mean_column

# 20-3)'sex','class','age' 기준으로 오름차순 정렬
groupby_mean_sort = groupby_mean_column.sort_values(['sex','class','age'], ascending = True)
groupby_mean_sort
```

```python
# 21. 범주형 변수의 도수분포표, 이원분할표
# 21-1) pandas를 활용한 도수분포표
import pandas as pd
df = pd.read_csv('data/titanic.csv')
df_survived = df['survived'].value_counts()
pd.DataFrame(df_survived, columns = ['survived'])

# 21-2) 기준이 두 개일 경우 > crosstab
df_cross_1 = pd.crosstab(df['class'], df['survived'])
print(df_cross_1)
print('----------')
df_cross_2 = pd.crosstab(df['class'], df['survived'], dropna = True) # dropna = True : NaN값 제외
print(df_cross_2)
print('----------')
# 21-3) 범주형 변수에 대한 전체 상대도수 비율
df_cross_3 = pd.crosstab(df['class'], df['survived'], normalize = True)
print(df_cross_3)
print('----------')
# 21-4) 범주형 변수에 대한 행 상대도수 비율
df_cross_4 = pd.crosstab(df['class'], df['survived'], normalize = 'index')
print(df_cross_4)
print('----------')
# 21-5) 범주형 변수에 대한 열 상대도수 비율
df_cross_5 = pd.crosstab(df['class'], df['survived'], normalize = 'columns')
print(df_cross_5)
```

```python
# 22. 데이터 재구조화 (melt, pivot_table)
import pandas as pd
import numpy as np
df = pd.read_csv('data/Fruits.csv')
print(df)
print('----------')

# 22-1) melt
df_melt = pd.melt(df, id_vars =['Fruit'], value_vars = ['Sales', 'Expenses', 'Profit'], var_name = 'subject', value_name = 'value', ignore_index = True)
print(df_melt)
print('----------')

# 22-2) pivot_table
# defaut : 자동 Group by + mean
# margins = True > 전체 평균 출력
df_pivot_mean = pd.pivot_table(df_melt, index = 'Fruit', columns = 'subject', values = 'value', margins = True) 
print(df_pivot_mean)
print('----------')
df_pivot_sum = pd.pivot_table(df_melt, index = ['Fruit'], columns = ['subject'], values = ['value'], aggfunc = ['sum'], margins = [True])
print(df_pivot_sum)
```

```python
# 24. 문자열 변경
import pandas as pd
import numpy as np
df = pd.read_csv('data/7.세종시_아파트(전월세)_실거래가.csv')
df.drop(columns = ['Unnamed: 0'], inplace = True)
df['전용면적(평)'] = df['전용면적(㎡)']/3.3

# 24-1) 앞 자리만 추출
df['시'] = df['시군구'].str[0:7] # 반열림구간
print(df)
print('----------')

# 24-2) 데이터 분할
# expand = True > 컬럼으로 분할, DataFrame 형식
# expand = False > 하나의 컬럼, Series 형식    
df_temp = df['시군구'].str.split(' ', expand = True) # expand = True > 컬럼으로 분할, DataFrame 형식
df = pd.concat([df, df_temp], axis = 1)
df.rename(
         columns = {
             0 : '시',
             1 : '군',
             2 : '구'
             
         },
         inplace = True
         )
print(df)
print('----------')
# 24-3) 문자열 포함 
# 이지더원(가락마을10단지), 이지더원(가락마을4단지) > 문자열 수정 필요
# 단지명 컬럼에 '가락마을' 찾은 후 수정 
array_temp = df[df['단지명'].str.contains('가락마을')]['단지명'].unique()
print(array_temp)
print('----------')
df.loc[df['단지명'] == '이지더원(가락마을10단지)', '단지명'] = '가락마을10단지(이지더원)'
df.loc[df['단지명'] == '이지더원(가락마을4단지)', '단지명'] = '가락마을4단지(이지더원)'
array_temp = df[df['단지명'].str.contains('가락마을')]['단지명'].unique()
print(array_temp)
print('----------')

# 24-5) unique()[시작 인덱스 : 끝 인덱스] # 반열림구간(끝 인덱스 미포함)
df.loc[df['단지명'] == '가락마을 20단지(호반베르디움5차)', '단지명'] = '가락마을20단지(호반베르디움5차)'
array_temp = df[df['단지명'].str.contains('가락마을')]['단지명'].unique()[0:4]
print(array_temp)
```
