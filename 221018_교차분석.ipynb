#### 교차분석 
##### > 범주형 척도로 구성된 자료 간의 연관관계를 확인하기 위해 교차표를 만들어 관계를 확인하는 방법
##### > 각 범주에 따른 결과변수의 분포 설명
##### > 카이제곱 통계량 사용 : (관찰빈도-기대빈도)의 제곱의 합 / 기대빈도
##### > 기대빈도 : 두 변수가 독립일 때 이론적으로 기대할 수 있는 빈도 분포

#### 1. 적합성 검정 
##### > 각 범주에 따르는 데이터의 관측빈도 분포가 이론적으로 기대빈도 분포를 따르는지 검정
##### > 모집단 분포에 대한 가설이 옳게 됬는지 관측치와 비교 검정
##### > 관측치들이 어떠한 이론적 분포를 따르는지 관찰 가능
##### > 자유도 : 범주형 변수 수 - 1 ex) 산, 바다의 선호도 > 2 -1 = 1
#### 2. 독립성 검정
##### > 모집단이 두 개의 변수 A,B에 의해 범주화 되었을 때, 각 범주가 독립인지 검정
##### > 자우도 : (i-1) * (j-1) ex) 지역에 따른 구매한다/안한다? > (2-1) * (2-1) = 1

#### 3. 동질성 검정
##### > 범주 A에 대해 범주 B의 유형들이 같은 빈도로 나타나는지 검정
##### > 독립성 검정이랑 동일하게 시행 하지만 관점의 차이


# 교차분석 과제 
## titanic데이터에서 sex와 class 간의 동질성 검정을 진행하세요. 

```python
import pandas as pd 
```

```python
data = pd.read_csv('data/titanic.csv')
data
```

#### 1. 가설설정
##### > H0 : sex는 class와 상관없이 동일하다.
##### > H1 : sex는 class에 따라 다르다.
#### 2. 유의수준 설정 : 0.05
#### 3. 기각역 설정 : 양측검정
#### 4. 전처리 및 통계량 계산(카이제곱 통계량)
##### > 범주 A에 대해 범주 B의 유형들이 같은 빈도로 분포되었는지 검정하기 위해, 카이제곱 통계량을 사용
#### 5. 해석 및 의사결정

```python
#4. 전처리 및 통계량 계산(카이제곱 통계량)
data.info() # sex, class는 결측치 X
```

```python
# 교차표 생성
table = pd.crosstab(data['sex'], data['class'])
table
```

```python
# 통계량 계산(카이제곱 통계량)
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table) # 카이제곱 통계량, 유의확률, 자유도, 기대빈도
result = "chi2 : {}, p-value : {}, df : {}".format(chi2, p, dof)
print(result)
print(expected)
```

##### 5. 해석 및 의사결정
###### > 카이제곱 통계량이 16.971 이며, 유의확률이 0.05 보다 작으므로 귀무가설을 기각하고 대립가설을 채택한다.
###### > 즉, class 관찰빈도는 sex 관찰빈도 수에 따라 다르게 나타난다.
