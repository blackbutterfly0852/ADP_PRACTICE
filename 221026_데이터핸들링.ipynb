```python
# 14. np.where(), np.select() 
df = pd.read_csv('data/titanic.csv')
df
```

```python
# 14-1) np.where(condition, True, False) : 데이터 프레임 혹은 리스트 내 필드에 접근하고 수정( == loc | iloc * 2)
# First class이면, temp컬럼에 'Yes' 반환 아니면 'No' 반환
import numpy as np
df['temp'] = np.where(df['survived'] == 1, 'Yes', 'No')
df.loc[df['survived'] == 1, 'temp2'] = 'Yes'
df.loc[df['survived'] == 0, 'temp2'] = 'No'
df
```

```python
# 14-2) np.select() : 데이터 프레임 혹은 리스트 내 필드에 접근하고 수정
# 조건 목록과 선택 목록을 입력으로 받아서 조건에 따라 선택 목록의 요소로 구성된 배열 반환
# 두 개이상의 조건이었을 때 Pandas의 주어진 조건을 기반으로 DataFrame의 열을 생성 가능.
condition_list = [
    (df['age'] >= 19) ,
    (df['age'] >= 8) & (df['age'] <19),
    (df['age'] < 8)]

choice_list = ['성인', '초중고등학생', '미취학아동']

df['status'] = np.select(condition_list, choice_list, default = 'Not_Specified') # default : 조건에 해당 X의 기본값

df
```

### 1. datetime 객체는 바로 str'f'time (문자열로), str'p'time(피나는 일자로) 가능 
##### ex) datetime.today().strftime('%Y %m %d %H %M %S %f')
##### ex) datetime.strptime(todaytime,'%Y %m %d %H %M %S %f')

### 2. Series 객체에서 datetime > str 경우 
##### ex) df[''].dt.strftime('날짜형식') 이용 # 형식은 원본과 동일하게
##### ex) df['datetime'].map(lambda x : datetime.strftime(x, '%Y-%m-%d'))

### 3. Series 객체에서 str > datetime 경우 
##### ex) pd.to_datetime(df[''], format = '날짜형식') 이용 # 형식은 원본과 동일하게
##### ex) df['str'].map(lambda x : datetime.strptime(x, '%Y-%m-%d'))

```python
# 28. 현재 날짜와 시간 데이터를 생성하고, 각각 다른 변수에 저장하여 데이터 형식을 확인하라
from datetime import datetime

today = datetime.today().strftime('%Y- dwd %m -dwd %d') # -dwd가 포함된 이유는 문자열임을 보여주기 위함.
print('today : ', today)
print('today 변수 type : ', type(today), '\n')

todaytime = datetime.today().strftime('%Y %m %d %H %M %S %f')
print('todaytime : ', todaytime)
print('todaytime 변수 type : ', type(todaytime), '\n')

todaytime_split = todaytime.split(" ")
print('today_time_split : ', todaytime_split)
print('today_time_split[0] : ', todaytime_split[0])
```

```python
# 29. str > datetime
datetime.strptime(todaytime,'%Y %m %d %H %M %S %f') # 위 todaytime과 동일한 날짜형식 필요
```

```python
# 30. 현재 시간을 구하고 이를 time 변수에 저장 후, 현재 시간의 연/월/일/시간/요일을 구하라
time = datetime.today()
print('년 : ', time.strftime('%Y'), ' & ', time.year, ' & ', time.strftime('%y'))
print('월 : ', time.strftime('%m'), ' & ', time.month)
print('일 : ', time.strftime('%d'), ' & ', time.day)
print('시 : ', time.strftime('%H'), ' & ', time.hour)
print('분 : ', time.strftime('%M'), ' & ', time.minute)
print('초 : ', time.strftime('%S'), ' & ', time.second)
print('밀리세컨드 : ', time.strftime('%f'), ' & ', time.microsecond)
print('요일 : ', time.strftime('%A'), ' & ', time.weekday(), ' & ', time.strftime('%a')) # 0 : 월요일, 6 : 일요일
```

```python
# 31. 현재 날짜로부터 100일 후의 날짜를 구하라
# datetime.timedelta(days=0, hours=0, minutes=0, seconds=0, microseconds=0, milliseconds=0,  weeks=0)
from datetime import timedelta
today = datetime.today()
after_one_hour = today + timedelta(hours = 1)
after_one_minute = today + timedelta(minutes = 1)
after_ten_second = today + timedelta(seconds = 10)
after_one_week = today + timedelta(weeks = 1)
after_total = today + timedelta(hours = 1, minutes = 1, seconds = 10, weeks = 1)
print('current time : ' , today)
print('after_one_hour : ' , after_one_hour)
print('after_one_minute : ', after_one_minute)
print('after_ten_second : ', after_ten_second)
print('after_one_week : ', after_one_week)
print('after_total : ', after_total)
```

```python
# 32. str > datetime
df = pd.read_csv('data/구매내역.csv')
df['datetime'] # int64
# int -> str 변환
df['datetime'] = df['datetime'].astype(str)
df['datetime'] # str
# str -> datetime 변환
df['to_datetime'] = pd.to_datetime(df['datetime'], format = '%Y%m%d') # 원본 데이터와 동일한 형식필요, 그러나 출력시 - 표시(ex, 20140627 > '%Y%m%d')
df
```

```python
# 33. datetime 형식인 컬럼에 특정 년도, 월, 일이 존재하는지?
print(df[df['to_datetime'].dt.year == 2014], '\n')
print(df[df['to_datetime'].dt.month == 7], '\n')
print(df[df['to_datetime'].dt.day == 7], '\n')

print(df[df['to_datetime'].dt.month.isin([5,6])])
```

```python
# 34. datetime > str , Series 형식은 'dt'가 필요
df['to_str'] = df['to_datetime'].dt.strftime('%Y-%m-%d')
df.info()
# str > datetime
# 원본 데이터와 동일한 형식필요, 그러나 출력시 - 표시(ex, 2014-06-27 > '%Y-%m%-d')
df['to_datetime_2'] = pd.to_datetime(df['to_str'], format = '%Y-%m-%d') 
df.info()
```

```python
# 35. map(lambda) 사용
# datetime > str
df['to_str_2'] = df['to_datetime_3'].map(lambda x : datetime.strftime(x, '%Y-%m-%d')) # str'f'time : 문자열로 변경

# str > datetime
df['to_datetime_3'] = df['to_str'].map(lambda x : datetime.strptime(x, '%Y-%m-%d')) # str'p'time : '피'보는 일자 암기

df.info()
df
```
