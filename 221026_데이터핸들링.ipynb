```python
# 5. 데이터프레임 요약
print(data.info()) # 데이터 크기, 타입, 결측치
print(data.describe()) # 데이터의 기초통계
print('데이터의 size : ', data['sepal length'].size) 
print('데이터의 shape : ', data['sepal length'].shape) # 튜플 형태로 행,렬 수를 반환
print('데이터의 min : ', data['sepal length'].min())
print('데이터의 max : ', data['sepal length'].max())
print('데이터의 unique : ',data['sepal length'].unique())

print('데이터의 mean : ', data['sepal length'].mean()) # NaN 제외
print('데이터의 count : ', data['sepal length'].count()) # NaN 제외
print('데이터의 value_counts() : ', data['sepal length'].value_counts()) # NaN 제외
```
