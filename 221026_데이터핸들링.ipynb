```python
# 14. np.where(), np.select() 
df = pd.read_csv('data/titanic.csv')
df
```

```python
# 14-1) np.where(condition, True, False) : 데이터 프레임 혹은 리스트 내 필드에 접근하고 수정( == loc | iloc * 2)
# First class이면, temp컬럼에 'Yes' 반환 아니면 'No' 반환
import numpy as np
df['temp'] = np.where(df['survived'] == 1, 'Yes', 'No')
df.loc[df['survived'] == 1, 'temp2'] = 'Yes'
df.loc[df['survived'] == 0, 'temp2'] = 'No'
df
```

```python
# 14-2) np.select() : 데이터 프레임 혹은 리스트 내 필드에 접근하고 수정
# 조건 목록과 선택 목록을 입력으로 받아서 조건에 따라 선택 목록의 요소로 구성된 배열 반환
# 두 개이상의 조건이었을 때 Pandas의 주어진 조건을 기반으로 DataFrame의 열을 생성 가능.
condition_list = [
    (df['age'] >= 19) ,
    (df['age'] >= 8) & (df['age'] <19),
    (df['age'] < 8)]

choice_list = ['성인', '초중고등학생', '미취학아동']

df['status'] = np.select(condition_list, choice_list, default = 'Not_Specified') # default : 조건에 해당 X의 기본값

df
```

```python
# 17. 데이터프레임 행/열 결합
customer1 = pd.DataFrame({'id':['c01', 'c02', 'c03', 'c04'],
                        'last_name':["Lee","Kim","Choi","Park"]},
                        index = [0,1, 2,3])
customer2 = pd.DataFrame({'Id':['c05','c06','c07'],
                         'Last_name':['Lim', 'Bae', 'Kim'],
                         'age':[23, 34, 45]},
                        index = [0, 1, 2])
# 데이터 프레임 행 결합
# customer1, customer2
axis_0 = pd.concat([customer1, customer2], axis = 0)
axis_0 # 컬럼명이 달라 기대한 결합과 다름

# 컬럼명 맞추기
customer1.columns = ['Id', 'Last_name']
#customer1.rename(columns = {'id' : 'Id', 'last_name' : 'Last_name'})

customer1

axis_0 = pd.concat([customer1, customer2], axis = 0, ignore_index = True) # ignore_index = True : 인덱스 재배열
axis_0

# 데이터 프레임 열 결합
axis_1 = pd.concat([customer1, customer2], axis = 1)
axis_1
```

```python
id_name = pd.DataFrame({'ID' : ['c01', 'c02', 'c03', 'c04', 'c05', 'c06', 'c07'],
                       'last_name': ['Lee', 'Kim', 'Choi', 'Park','Lim','Bae','Kim']})
id_number = pd.DataFrame({'id' : ['c03', 'c04', 'c05', 'c06', 'c07', 'c08', 'c09'],
                         'number' : [3, 1, 0, 7, 3, 4, 1]})
```

```python
id_name
```

```python
id_number.rename(columns = {'id':'ID'}, inplace = True)
id_number
```

```python
pd.merge(id_name, id_number, how = 'inner', on = 'ID') # inner join
```

```python
pd.merge(id_name, id_number, how = 'left', on = 'ID') # left outer join
```

```python
pd.merge(id_name, id_number, how = 'right', on = 'ID') # right outer join
```

```python
pd.merge(id_name, id_number, how = 'outer', on = 'ID') # full outer join
```

```python
# 20. 특정 컬럼을 기준으로 데이터를 그룹지어 집계함수 적용
import pandas as pd
df = pd.read_csv('data/titanic.csv')

df.info()
# 20-1) 'sex','class' 을 기준으로 데이터를 그룹지어 평균함수 적용 > 숫자형 컬럼 제외 추출
df.groupby(['sex', 'class']).mean()
```

```python
# 20-2) 20-1) + 인덱스 리셋
df.groupby(['sex', 'class']).mean().reset_index()
```

```python
# 20-3) 'sex','class' 을 기준으로 데이터를 그룹지어 평균함수 적용 후 'age' 와 'fare' 추출
df_gb = df.groupby(['sex','class']).mean()[['age','fare']]
df_gb
```

```python
# 20-4) 20-3) + 'sex','class','age' 기준으로 오름차순 정렬
df_gb.sort_values(['sex', 'class','age'], ascending = False)
```

```python
# 21) 범주형변수의 도수분포표, 이원분할표
# 21-1) pandas를 활용한 도수분포표
class_count = df['survived'].value_counts()
pd.DataFrame(class_count)
```

```python
# 21-2) 기준이 두 개일 경우 > crosstab
pd.crosstab(df['class'],df['survived'], dropna = True) # NaN 값 제거
```

```python
# 21-3) 범주형 변수에 대한 상대도수 비율
pd.crosstab(df['class'], df['survived'], normalize = True) # 행 / 열 포함
# print(0.089787 + 0.152637 + 0.108866 + 0.097643 + 0.417508 + 0.133558) # = 0.99999
```
