### 정규화(Regularization)
#### 1. What is a good model? Explanatoy modeling + Predictive modeling
##### 1) Explanatoy Modeling : 현재 데이터를 얼마나 잘 설명하는가? > Training Error 최소 == MSE 최소
##### 2) Predictive Modeling : 미래 데이터에 대해 얼마나 잘 예측하는가? Expected Error(Testing Error) 최소

#### 2. Expected MSE  = Irreducible Error + Bias^2(편향) + Variance(변동) 
##### 1) Low Bias, Low Variance / High Bias, Low Variance / Low Bias, High Variance / High Bias, High Variance > Low 일수록 좋다.
##### 2) 과녁의 정중앙은 Test Data에 대한 정답
##### 3) 성능상 1, 4 번은 좋고 않좋음이 분명하지만, 2,3번은 상황에 따라 다르다.

#### 3. BLUE( Best Linear Unbiased Estimator ) : 최소제곱법으로 계산된 회귀계수는 편향되지 않는 점추정량 
##### 1) Biased 관점으로만 해석.
##### 2) Unbiased Estimator 중 Variance가 가장 작은 모델을 선택하는 과정

#### 4. Unbiased를 약간 희생하더라도(약간 Biased 하게 해도) Variance(분산)을 감소시키면 미래에 대한 예측이 더 좋아지지 않을까?
##### 1) SubSet Selection : 전체 p개의 설명변수(X) 중 일부 k개만 사용하여 회귀계수 B추정
##### 2) 1차, 2차, 4차식에 대한 차수(X축)/오류(Y축) 
##### 3) 1차일 때는 Training Data와 Testing Data 오류가 높다. (High Bias, Low Variance), 2번(과녁의 정중앙 옆에 모여)
##### 4) 4차일 때는 Training Data의 오류는 낮고, Testing Data의 오류는 다시 높아진다. (Low Bias, High Variance), 3번 (과녁의 정중앙 방향이지만 퍼져)
##### 5) 2차식은 Bias를 약간 희생하고 Variance를 감소시키기 때문에 현재 데이터 설명력뿐만 아니라 미래 데이터 예측도 적합한 모델

#### 5. 정규화 : Least Square Solution으로 Training Data에 overfitting(과적합) 모델을 만들었을 때  B(회귀계수)에 제약을 주어 정돈한다는 의미
#### 6. Ridge Regression
##### 1) B(회귀계수)의 L2(제곱) normalization을 제약 
##### 2) 추정된 릿지 B값이 최소가 되기 위해? 람다가 커지면 B값(B의 영향력)작아지고, 람다가 작아지면 B값(B의 영향력)이 커진다.(0에 가까워져야 한다.)
##### 3) 람다가 커진다? (is very Big) : B를 크게 제약하며, (Training Data) High Biased, Low Variance > 현재 데이터에 대한 설명력이 떨어진다.(UnderFitting)
##### 4) 람다가 작아진다? (is very Small) : B의 제약이 거의 없다, (Testing Data) Low Biased, High Variance > 미래 데이터에 대한 예측력이 떨어진다(OverFitting)
##### 5) 람다(Hyper Parameter, 사용자 결정)을 통해 3), 4)의 상충관계를 잘 조절 필요
##### 6) Least Square Method : B가 무슨 값을 가져도 좋으니 MSE 값만 최소화(Biased 관점)
##### 7) Regulation Method : MSE + Variance 고려, LSM보다는 Biased 하겠지만 미래 데이터에 대한 예측력은 향상
##### 8) B1, B2의 제약조건(원) + MSE 타원 그림(2->3), penalty(U 형태), closed form solution

#### 7. LASSO(Least Absolute 'Shrinkage' and 'Selection' Operator)
##### 1) Shrinkage(회귀계수 파라미터가 값이 작아진다), Selection(Y값 예측에 중요하지 않은 변수 선택을 통해 중요한 변수를 알 수 있다.)
##### 2) SUM(|B|) <=  T, 람다가 커지면 (T값이 작아지면) B에 대한 제약을 크게하며, 람다가 작아지면(T값이 커지면) B에 대한 제약을 거의 없게 한다.
##### > 람다가 커질수록 B값들이 작아진다(Shrinkage, 0으로 수렴)
##### > 람다가 커진다? : 점점 Least Square Solution에 가까워지며, 적은 변수, 간단한 모델, 해석 쉬움, 높은 학습 오차(Training Data에 대한 Under Fitting)
##### > 람다가 작아진다? : 많은 변수, 복잡한 모델, 해석 어려움, 낮은 학습 오차(Training Data에 대한 Overfitting)
##### 3) B1, B2의 제약조건(마름모) + MSE 타원 그림(2->3), penalty(V 형태), numerical Optimization (closed form solution X)
##### 4) 마름모이기 때문에 꼭지점이 존재하며, 이는 Y를 예측하는데 중요하지 않는 변수로 판단.

#### 8. Robust Model(강건한 모델)
##### 1) 데이터가 달라질 때마다 변수 선택이 달라지지 않을까? 변화가 작다면 Robust Model(강건한 모델)이라고 할 수 있다.
##### 2) LASSO의 경우 B값들의 박스플롯 범위가 작다 > Robust Model이라고 할 수 있다.

#### 9. Solution Path
##### 1) Ridge : Shrinkage(람다값이 커질수록, T값이 작아질 수록 B값은 작아진다) vs  LASSO : Shrinkage + Selection
##### 2) Ridge : B값이 작아지지만 0은 되지 않는다. 변수 간 상관관계 높은 상황에서 예측 성능이 좋다.
##### 3) LASSO : B값이 0이 될 수 있다. 변수 간 상관관계가 높은 상황에서 Ridge보다 상대적으로 예측 성능이 떨어진다.
###### > LASSO의 한계점 변수 간 상관관계가 높을 경우 B값의 박스플롯의 범위가 넒어진다.

#### 10. Elastic Modeling
##### 1) LASSO 모델에 Ridge 추가 제약
##### 2) Elastic으로 추정된 B1과 B2의 상관관계가 1이면 이는 Y값을 예측하는데 동일하게 중요한/중요하지 않는 변수이다. (Grouping Effect)
##### 3) Grid Search를 통해 람다1, 람다2를 찾아서 에러가 가장 작은 B값들을 추정
##### 4) B1, B2의 제약조건(원X, 마름모X) + MSE 타원 그림(2->3), penalty(Ridge와 Lasso의 중간 형태)

#### 11. MAE(Mean Absolute Error) vs RMSE(Root Mean Squared Error)
##### 1) MAE : 현실적이고 직관적인 에러
##### 2) RMSE : 크기에 의존적인 에러
##### 3) 매우 큰 에러로 오분류가 된 데이터가 들어오면 RMSE가 해당 데이터에 큰 패널티를 부여
##### 4) 예측 모델이 데이터를 학습할 때 에러가 큰 데이터(특이값)에 민감하지 않고 잘 학습하기 위함(Scaled-Dependet Error, 에러에 대한 Robust한 Metric)

```python
import pandas as pd 
df = pd.read_csv("data/kc_house_data.csv")
df
```

```python
# 1. 전처리
# 1-1) 결측치 확인 
df.info() # 결측치 X
```

```python
# 1-2) id, date 삭제
df.drop(columns = ['id', 'date'], inplace = True)
df
```

```python
# 1-3) 범주형 변수 전처리
print(df['waterfront'].value_counts())
df = pd.get_dummies(df, columns = ['waterfront', 'view', 'condition','grade'])
df
```

```python
# 1-4) 종속변수 y = price, 나머지는 독립변수 X로 설정
y = df['price']
X = df.drop(columns = ['price'])
```

```python
# 1-5) 홀드아웃 기법을 사용하여 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1000)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# 2. Ridge 회귀분석
import warnings
warnings.filterwarnings(action = 'ignore')

from sklearn.linear_model import Ridge
import numpy as np

ridge = Ridge()
ridge.fit(X_train, y_train)

print("사용한 컬럼의 수 : " , np.sum(ridge.coef_ != 0)) # Ridge의 계수, 0이 아니면 해당 컬럼을 사용
print("Train 데이터 성능 : ", ridge.score(X_train, y_train))
print("Test 데이터 성능 : ", ridge.score(X_test, y_test))
```

```python
# 2-1) RMSE
y_pred = ridge.predict(X_test)
resid = y_test - y_pred
rmse = np.sqrt(sum(resid**2)/len(y_test))
print("RMSE : " , rmse) # 

from sklearn.metrics import mean_squared_error
print("RMSE : " , np.sqrt(mean_squared_error(y_test, y_pred)))
```
