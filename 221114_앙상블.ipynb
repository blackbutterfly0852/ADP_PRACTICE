# 13. 앙상블과제 


## boston데이터를 가지고 집값을 예측하는 아래 3가지의 모델을 만들고 모델을 비교하고 최적 모델의 결과를 해석하세요

### 배깅
### 랜덤포래스트
### 부스팅 



```python
import pandas as pd 
from sklearn.datasets import load_boston
boston= load_boston()
X = pd.DataFrame(boston.data,columns=boston.feature_names)
y = boston.target
```

```python
X
```

```python
y
```

```python
#0. 전처리
from sklearn.model_selection import train_test_split
X.info() # 결측치 X
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 2022)

# 1. 의사결정나무
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np


decisionTreeRegression = DecisionTreeRegressor()
param_grid = {'max_depth' : range(1,40,1), 'min_samples_split' : range(2,40,1)}
model = GridSearchCV(decisionTreeRegression, param_grid)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('OneDecisionTree_best_estimator : ', model.best_estimator_)
print('OneDecisionTree_train_score : ', model.score(X_train, y_train))
print('OneDecisionTree_test_score : ', model.score(X_test, y_test))
print('OneDecisionTree_RMSE : ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# 2. Bagging
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import GridSearchCV

baggingRegressor = BaggingRegressor(oob_score = True)
param_grid = {'n_estimators' : range(10,100,10)}
baggingRegressor = GridSearchCV(baggingRegressor, param_grid)
baggingRegressor.fit(X_train, y_train)

print('Bagging_get_params : ' , baggingRegressor.get_params)
print('Bagging_best_estimator : ', baggingRegressor.best_estimator_)
print('Bagging_best_estimator_.n_estimators : ', baggingRegressor.best_estimator_.n_estimators)

# oob_score 호출을 위해 Bagging 객체 재생성
model = BaggingRegressor(oob_score = True, n_estimators = baggingRegressor.best_estimator_.n_estimators )
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Bagging_train_score : ', model.score(X_train, y_train))
print('Bagging_test_score : ', model.score(X_test, y_test))
print(f"Bagging_oob_score : , {model.oob_score_}")
print('Bagging_RMSE : ', np.sqrt(mean_squared_error(y_test, y_pred)))

print('Bagging_feature_importance')
feature_importances = pd.DataFrame(np.mean([ tree.feature_importances_ for tree in model.estimators_], axis = 0))
feature_importances = pd.concat([pd.DataFrame(X.columns),feature_importances], axis = 1)
feature_importances.columns = ['var_name', 'var_value']
feature_importances = feature_importances.sort_values(['var_value'], ascending = False).reset_index(drop=True)
print(feature_importances)

import matplotlib.pyplot as plt
plt.barh(feature_importances['var_name'], feature_importances['var_value'], align = 'center')
plt.xlabel('feature_importance', size = 15)
plt.ylabel('feature', size = 15)
plt.show()
```

```python
# 3. 랜덤포레스트
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

randomForestRegressor = RandomForestRegressor(oob_score = True)
param_grid = {'n_estimators' : range(10,100,10)}
randomForestRegressor = GridSearchCV(randomForestRegressor, param_grid)
randomForestRegressor = randomForestRegressor.fit(X_train, y_train)

print('RandomForest_get_params : ', randomForestRegressor.get_params())
print('RandomForest_best_estimator : ', randomForestRegressor.best_estimator_)
print('RandomForest_best_estimator_.n_estimators : ', randomForestRegressor.best_estimator_.n_estimators)

# oob_score 호출을 위해 Bagging 객체 재생성
model = RandomForestRegressor(oob_score = True, n_estimators = randomForestRegressor.best_estimator_.n_estimators)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print('RandomForest_train_score : ', model.score(X_train, y_train))
print('RandomForest_test_score : ', model.score(X_test, y_test))
print(f'RandomForest_oob_score : {model.oob_score_}')
print('RandomForest_RMSE : ', np.sqrt(mean_squared_error(y_test, y_pred)))

print('RandomForest_feature_importance')
feature_importances = pd.DataFrame(np.mean([tree.feature_importances_ for tree in model.estimators_], axis = 0))
feature_importances = pd.concat([pd.DataFrame(X.columns), feature_importances], axis = 1)
feature_importances.columns = ['var_name', 'var_value']
feature_importances = feature_importances.sort_values(['var_value'], ascending = False).reset_index(drop = True)

import matplotlib.pyplot as plt
plt.barh(feature_importances['var_name'], feature_importances['var_value'], align = 'center')
plt.xlabel('feature_importances', size = 15)
plt.ylabel('feature', size = 15)
plt.show()
    
```

```python
# 4. 부스팅
# 회귀계수에 영향을 많이 주는 변수만 추출, 발전 > 영향을 미치는 변수와 아닌 변수간 (성능) 차이 존재 > 마지막 성능을 높이고자 사용
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
import numpy as np
adaBoostRegressor = AdaBoostRegressor()
# learning_rate : 클수록 과소적합, 작을수록 과대적합
param_grid = {'learning_rate' : np.arange(0.01, 0.1, 0.02)} # 0.01 ~ 0.1 이전까지 0.02 씩 증가
model = GridSearchCV(adaBoostRegressor, param_grid)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('AdaBoost_best_estimator : ', model.best_estimator_)
print('AdaBoost_train_score : ', model.score(X_train, y_train))
print('AdaBoost_test_score : ', model.score(X_test, y_test))
print('AdaBoost_RMSE : ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('AdaBoost_feature_importance')
feature_importances = pd.DataFrame(np.mean([tree.feature_importances_ for tree in model.best_estimator_.estimators_], axis = 0))
feature_importances = pd.concat([pd.DataFrame(X.columns), feature_importances], axis = 1)
feature_importances.columns = ['var_name', 'var_value']
feature_importances = feature_importances.sort_values(['var_value'], ascending = False).reset_index(drop = True)

import matplotlib.pyplot as plt
plt.barh(feature_importances['var_name'], feature_importances['var_value'], align = 'center')
plt.xlabel('feature_importances', size = 15)
plt.ylabel('feature', size = 15)
plt.show()
```

## titanic데이터를 가지고 생존율을 예측하는  아래 3가지의 모델을 만들고 모델을 비교하고 최적 모델의 결과를 해석하세요

### 배깅
### 랜덤포래스트
### 부스팅 



```python
df = pd.read_csv("data/titanic.csv")
df
```

```python
# 0. 전처리
import pandas as pd
from sklearn.model_selection import train_test_split

# 결측치 대치 필요: age, embarked(embark_town)
df.info() 
df['age'].fillna(df['age'].mean(), inplace = True)
df['embarked'].fillna(df['embarked'].value_counts().index[0], inplace = True)

# 중복컬럼삭제 필요
df.drop(['sex','class','embark_town'], axis = 1, inplace = True)

# 더미변수화 필요
df = pd.get_dummies(df,columns = ['embarked']) 

# 데이터 셋 분할
X = df.drop(columns = ['survived'])
y = df['survived']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 2022)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# 1. 의사결정나무
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, classification_report
from sklearn.metrics import plot_roc_curve, roc_auc_score

decisionTreeClassifier = DecisionTreeClassifier()
param_grid = {'max_depth' : range(1,40,2), 'min_samples_leaf' : range(2,40,2)}
model = GridSearchCV(decisionTreeClassifier, param_grid)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('DecisionTreeClassifier_best_estimator : ', model.best_estimator_)
print('DecisionTreeClassifier_train_score : ', model.score(X_train,y_train))
print('DecisionTreeClassifier_test_score : ', model.score(X_test,y_test))

print('DecisionTreeClassifier_confusion_matrix')
print( pd.DataFrame(confusion_matrix(y_test, y_pred), index = ['True[0]', 'True[1]'], columns = ['Pred[0]', 'Pred[1]']))
print('DecisionTreeClassifier_accuracy_score : ', accuracy_score(y_test, y_pred))
print('DecisionTreeClassifier_recall_score : ', recall_score(y_test, y_pred))
print('DecisionTreeClassifier_precision_score : ', precision_score(y_test, y_pred))
print('DecisionTreeClassifier_classification_report')
print(classification_report(y_test, y_pred))
# ROC_SCORE
# bineary 분류일 때 : roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1])
# 다중 클래스 분류일 때 : roc_auc_score(y_test, clf.predict_proba(X_test), multi_class='ovr')
# 여러 개의 라벨을 분류할 때 : roc_auc_score(y, y_pred, average=None)
print('DecisionTreeClassifier_ROC_AUC_SCORE : ', roc_auc_score(y_test, model.predict_proba(X_test)[:,-1]))
# ROC 곡선 (X축 : 민감도, Y축 : 1-특이도(실제 False를 False라고 예측하는 비율))
print('DecisionTreeClassifier_ROC_CURVE') 
print(plot_roc_curve(model, X_test, y_test))

# 변수 중요도 확인
```

```python
# 2. Bagging
from sklearn.ensemble import BaggingClassifier # base_estimator default = DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, classification_report
from sklearn.metrics import plot_roc_curve, roc_auc_score
import matplotlib.pyplot as plt

baggingClassifier = BaggingClassifier(oob_score = True)
param_grid = {'n_estimators' : range(10,100,10)}
baggingClassifier = GridSearchCV(baggingClassifier, param_grid)
baggingClassifier.fit(X_train, y_train)
print('Bagging_get_params : ', baggingClassifier.get_params)
print('Bagging_best_estimator_ : ', baggingClassifier.best_estimator_)
print('Bagging_best_estimator_.n_estimators : ', baggingClassifier.best_estimator_.n_estimators)

model = BaggingClassifier(oob_score = True, n_estimators = baggingClassifier.best_estimator_.n_estimators)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('Bagging_train_score : ', model.score(X_train,y_train))
print('Bagging_test_score : ', model.score(X_test, y_test))
print('Bagging_confusion_matrix')
print(pd.DataFrame(confusion_matrix(y_test, y_pred), index = ['True[0]', 'True[1]'], columns = ['Pred[0]', 'Pred[1]']))
print('Bagging_accuracy_score : ', accuracy_score(y_test, y_pred))
print('Bagging_recall_score : ', recall_score(y_test, y_pred))
print('Bagging_precision_score : ', precision_score(y_test, y_pred))
print('Bagging_classification_report')
print(classification_report(y_test, y_pred))
print('Bagging_ROC_AUC_SCORE : ', roc_auc_score(y_test, model.predict_proba(X_test)[:,-1]))
print('Bagging_ROC_CURVE')
plot_roc_curve(model, X_test, y_test) # 파라미터 주의
```

```python
# 변수 중요도
feature_importance = pd.DataFrame(np.mean([tree.feature_importances_ for tree in model.estimators_], axis = 0))
feature_importance = pd.concat([pd.DataFrame(X.columns), feature_importance], axis = 1)
feature_importance.columns = ['var_name', 'var_value']
feature_importance = feature_importance.sort_values(['var_value'], ascending = False)
print(feature_importance)
import matplotlib.pyplot as plt
plt.barh(feature_importance['var_name'], feature_importance['var_value'], align = 'center')
plt.xlabel('feature importance', size = 15)
plt.ylabel('feature', size = 15)
```
